```python
plt.figure(figsize=(12,12))
#plt.plot(df['Year'],df['china'],color='red')
plt.plot(df['Year'],df['USA'],color='blue',label='USA')
plt.plot(df['Year'],df['china'],color='red',label='China')
plt.plot(df['Year'],df['France'],color='green',label='France')
plt.plot(df['Year'],df['Germany'],color='orange',label='Germany')
plt.plot(df['Year'],df['Norway'],color='Yellow',label='Norway')
plt.xlabel("Years",fontsize=20)
plt.ylabel('Sales of Electric Vechicles',fontsize=20)
plt.title("Sales of Electric Vechicles in Different countries",fontsize=20)
plt.grid()
plt.legend(fontsize=20)
plt.savefig('D:\my_plot.png')
```
